```python
%%sql
CREATE TABLE Movies(
	title 				CHAR(100) DEFAULT '?',
	year 				INT,
	length				INT,
	genre				CHAR(10),
	studioName			CHAR(30),
	studioAddress 	    CHAR(50)
)
```

```python
%sqlcmd columns --table Movies
```

```python
%%sql
ALTER Table Movies ADD producerPhone CHAR(16) DEFAULT 'unlisted';
```

```python
%%sql
CREATE TABLE Movies(
	title 		CHAR(100),
	year 			INT,
	length		    INT,
	genre			CHAR(10),
	studioName		CHAR(30),
	producerCNum 	INT,
	PRIMARY KEY 	(title, year)
	
);
```

### Single Table Queries

```python
%%sql
CREATE TABLE Students (
	sid		   	INT,
	name        CHAR(20),
	school  	CHAR(20),
	age			INTEGER,
	gpa 		REAL
)
```

Inserting multiple rows into Students, in preparation for the next set of queries:

```python
%%sql
INSERT INTO Students (sid, name, school, age, gpa)
VALUES 
    (1001, 'Adam', 'SFU', 23, 3.2),
    (1002, 'Aiden', 'UBC', 19, 3.5),
    (1003, 'Alice', 'SFU', 18, 3.7),
    (1004, 'Bob', 'UBC', 22, 3.1),
    (1005, 'David', 'SFU', 20, 3.2),
    (1006, 'John', 'SFU', 21, 3.1),
    (1007, 'Mary', 'UBC', 21, 3.4),
    (1008, 'Mike', 'SFU', 24, 3.1),
    (1009, 'Sarah', 'UBC', 18, 3);
```

### NULL Values

```python
%%sql
Drop Table Students
```

```python
%%sql
CREATE TABLE Students (
	sid	    INT,
	name    CHAR(20) NOT NULL,
	school  CHAR(20) NOT NULL,
	age	    INTEGER,
	gpa 	FLOAT
);
```

```python
%%sql
INSERT INTO Students (sid, name, school, age, gpa)
VALUES 
    (1001, 'Adam', 'SFU', 23, 3.2),
    (1002, 'Aiden', 'UBC', 19, 3.5),
    (1003, 'Alice', 'SFU', 18, 3.7),
    (1004, 'Bob', 'UBC', 22, 3.1),
    (1005, 'David', 'SFU', 20, NULL),
    (1006, 'John', 'SFU', 21, 3.1),
    (1007, 'Mary', 'UBC', 21, 3.4),
    (1008, 'Mike', 'SFU', 24, 3.1),
    (1009, 'Sarah', 'UBC', 18, 3);

SELECT * FROM Students;
```

```python
%%sql
SELECT gpa*100 FROM students
```

```python
%%sql
SELECT gpa*0 FROM students
```

```python
%%sql
SELECT name FROM students WHERE gpa > 3.5
```

```python
%%sql
SELECT name FROM students WHERE age > 19 OR gpa > 3.5
```

```python
%%sql
SELECT * FROM students WHERE age > 15 OR gpa > 3.5
```

### Checking NULL Values

```python
%%sql
SELECT name FROM students WHERE gpa = NULL
```

```python
%%sql
SELECT name FROM students WHERE gpa is NULL
```

```python
%%sql
SELECT * FROM   Students 
WHERE  age < 25 OR age >= 25
```

```python
%%sql
SELECT * FROM   Students 
WHERE  age < 25 OR age >= 25 OR age IS NULL
```
