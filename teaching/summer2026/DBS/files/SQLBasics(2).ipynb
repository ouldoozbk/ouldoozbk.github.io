## Foreign Keys

```python
%%sql 
CREATE TABLE Students(
  sid 	      CHAR(20),
  name        CHAR(20),
  school      CHAR(15),
  age         INT,
  gpa	      DOUBLE,
  PRIMARY KEY (sid)
);
```

```python
%%sql
CREATE TABLE Enrolled(
	stid 	CHAR(20),
	cid	    CHAR(20),
	grade	CHAR(5),
	PRIMARY KEY (stid, cid),
	FOREIGN KEY (stid) REFERENCES   Students(sid)
)
```

### Joining Tables

```python
%%sql
SELECT name 
FROM Students, Enrolled 
WHERE sid=stid AND cid=125 AND grade='A+'
```

```python
%%sql
SELECT name FROM Students Join Enrolled ON sid=stid 
WHERE  cid=125 AND grade='A+'
```

```python
%%sql
SELECT name 
FROM Students 
JOIN Enrolled ON sid=stid 
    AND cid=125 AND grade='A+'
```

### Set Operations

```python
%%sql
SELECT name 
FROM   Students S, Enrolled E1, Enrolled E2
WHERE  S.sid = E1.stid AND S.sid = E2.stid 
       AND (E1.cid = 120 AND E2.cid = 125)
```

```python
%%sql
SELECT name
FROM   Students, Enrolled
WHERE  sid = stid AND cid = 120
UNION
SELECT name
FROM   Students, Enrolled
WHERE  sid = stid AND cid = 125
```

```python
%%sql
SELECT name
FROM   Students, Enrolled 
WHERE  sid = stid AND cid = 120
INTERSECT
SELECT name
FROM   Students, Enrolled
WHERE  sid = stid AND cid = 125
```

```python
%%sql
SELECT name
FROM   Students, Enrolled 
WHERE  sid = stid AND cid = 125
EXCEPT
SELECT name
FROM   Students, Enrolled
WHERE  sid = stid AND cid = 120
```

## Inner, Outer, and Cross Joins

```python
%%sql
INSERT INTO Students (sid, school, name, age)
VALUES (1015, 'UBC', 'Michael', 21)
```

```python
%%sql
INSERT INTO Enrolled (cid, stid, grade)
VALUES (354, 1015, 'A+')
```

```python
%%sql
SELECT name, cid, grade 
FROM Students CROSS JOIN Enrolled ON sid = stid
```

```python
%%sql 
SELECT name, cid, grade 
FROM Students INNER JOIN Enrolled ON sid = stid
```

```python
%%sql 
SELECT name, cid, grade 
FROM   Students FULL OUTER JOIN Enrolled ON sid = stid
```

```python
%%sql 
SELECT name, cid, grade 
FROM   Students LEFT OUTER JOIN Enrolled ON sid = stid
```

```python
%%sql 
SELECT name, cid, grade 
FROM  Students RIGHT OUTER JOIN Enrolled ON sid = stid
```
