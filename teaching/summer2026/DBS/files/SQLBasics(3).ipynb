### Preparing The Tables

```python
%%sql 
CREATE TABLE Students(
  sid 	      CHAR(20),
  name        CHAR(20),
  school      CHAR(15),
  age         INT,
  gpa	      DOUBLE,
  PRIMARY KEY (sid)
);
CREATE TABLE Enrolled(
	stid 	CHAR(20),
	cid	    CHAR(20),
	grade	CHAR(5),
	PRIMARY KEY (stid, cid),
	FOREIGN KEY (stid) REFERENCES   Students(sid)
);
INSERT INTO Students (sid, name, school, age, gpa)
VALUES 
    (1001, 'Adam', 'SFU', 23, 3.2),
    (1002, 'Aiden', 'UBC', 19, 3.5),
    (1003, 'Alice', 'SFU', 18, 3.7),
    (1004, 'Bob', 'UBC', 22, 3.1),
    (1005, 'David', 'SFU', 20, 3.2),
    (1006, 'John', 'SFU', 21, 3.1),
    (1007, 'Mary', 'UBC', 21, 3.4),
    (1008, 'Mike', 'SFU', 24, 3.1),
    (1009, 'Sarah', 'UBC', 18, 3);
INSERT INTO Enrolled (stid, cid, grade)
VALUES 
        (1001, 200, 'A'),
        (1001, 295, 'A'),
        (1001, 250, 'B+'),
        (1002, 130, 'A'),
        (1002, 125, 'A+'),
        (1003, 120, 'A'),
        (1003, 125, 'B'),
        (1003, 150, 'B'),
        (1004, 354, 'A');
```

### Conditions involving relations

```python
%%sql

/* An alternative to the previous query */
SELECT name 
FROM Students 
WHERE sid in 
	(	SELECT stid
        FROM Enrolled
        WHERE grade = 'A' 
	)
```

```python
%%sql
SELECT name 
FROM Students 
WHERE sid not in 
	(	SELECT stid
        FROM Enrolled
        WHERE grade = 'A' 
	)
```

```python
%%sql
SELECT DISTINCT name 
FROM Students, ( SELECT stid
                 FROM Enrolled
                 WHERE grade = 'A' 
            	) as Grades
WHERE sid = Grades.stid
```

```python
%%sql
SELECT DISTINCT name 
FROM Students, ( SELECT stid
                 FROM Enrolled
                 WHERE grade = 'A' 
            	) 
WHERE sid = stid
```

### Conditions involving tuples

```python
%%sql
SELECT stid
FROM Enrolled 
WHERE (stid, cid) IN
      (
        SELECT stid, cid
        FROM Enrolled 
        WHERE  grade = 'A'
      )
```

```python
%%sql
SELECT name 
FROM Students 
WHERE sid IN (   
                 SELECT stid
                 FROM Enrolled 
                 WHERE (stid, cid) IN
                    (
                        SELECT stid, cid
                        FROM Enrolled 
                        WHERE  grade = 'A'
                    )
             ) 
```
